```python
from ultralytics import YOLO
import cv2

# Detection model initialize
detection = YOLO("/home/maverick/Track1_2025/data/Warehouse_002/yolo11n.pt")

# Specify your video path
video_path = "data/Warehouse_013/Camera_01.mp4"

# Set confidence threshold and classes
conf_threshold = 0.5  # Adjust this value between 0 and 1
classes = [0]  # 0 is the class index for 'person' in COCO dataset

# Run inference on video
results = detection(
    video_path,
    stream=True,
    conf=conf_threshold,
    classes=classes  # Only detect persons
)

# Process each frame
for r in results:
    # Get the frame with predictions
    frame = r.plot()
    
    # Display the frame (press 'q' to quit)
    cv2.imshow('Person Detection', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Clean up
```
